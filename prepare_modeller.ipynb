```python
from Bio import PDB

# Ruta al archivo PDB
pdb_file = "5is5-completo.pdb"

# Crear un objeto de estructura PDB
parser = PDB.PDBParser()
structure = parser.get_structure("5si5", pdb_file)

# Iterar sobre todas las cadenas presentes
for model in structure:
    for chain in model:
        print("Cadena:", chain.id)

        # Iterar sobre todos los residuos presentes en la cadena
        for residue in chain:
            # Verificar si el residuo no es agua ni un ion
            if residue.get_resname() != "HOH" and residue.get_resname() != "WAT" and residue.get_resname() != "ION":
                aminoacid_name = residue.get_resname()
                aminoacid_number = residue.get_id()[1]
                print("El primer aminoácido es:", aminoacid_name, "Número de secuencia:", aminoacid_number)
                break

        break  # Mostrar solo la primera cadena

    break  # Mostrar solo el primer modelo
```

```python
import pandas as pd
from Bio import PDB

# Ruta al archivo PDB
pdb_file = "5is5-completo.pdb"

# Crear un objeto de estructura PDB
parser = PDB.PDBParser()
structure = parser.get_structure("5is5", pdb_file)

# Lista de aminoácidos
aminoacids = []

# Iterar sobre todas las cadenas presentes
for model in structure:
    for chain in model:
        #print("Cadena:", chain.id)

        # Iterar sobre todos los residuos presentes en la cadena
        for residue in chain:
            residue_number = residue.get_id()[1]
            aminoacid_name = residue.get_resname()  # Obtener el nombre del aminoácido
            aminoacids.append({ "Aminoacids": aminoacid_name,"ID": residue_number,})

        break  # Mostrar solo la primera cadena

    break  # Mostrar solo el primer modelo

# Crear DataFrame
df_pdb = pd.DataFrame(aminoacids)
```

```python
import pandas as pd

# Ruta al archivo PDB
pdb_file = "5is5-completo.pdb"

# Leer el archivo PDB y buscar las líneas con información de residuos faltantes
missing_residues = []
with open(pdb_file, "r") as file:
    for line in file:
        if line.startswith("REMARK 465"):
            residue_info = line.split()[2:5]
            missing_residues.append(residue_info)

# Crear el DataFrame con los datos de los residuos faltantes
df = pd.DataFrame(missing_residues, columns=["Aminoacids", "Cadena", "ID"])
df = df[7:]
# Imprimir el DataFrame
df
```

```python
aminoacid_codification = {
    'ALA': 'A',
    'ARG': 'R',
    'ASN': 'N',
    'ASP': 'D',
    'CYS': 'C',
    'GLN': 'Q',
    'GLU': 'E',
    'GLY': 'G',
    'HIS': 'H',
    'ILE': 'I',
    'LEU': 'L',
    'LYS': 'K',
    'MET': 'M',
    'PHE': 'F',
    'PRO': 'P',
    'SER': 'S',
    'THR': 'T',
    'TRP': 'W',
    'TYR': 'Y',
    'VAL': 'V'
}
```

```python
# Convertir columna "ID" a tipo int
df['ID'] = df['ID'].astype(int)
df_pdb['ID'] = df_pdb['ID'].astype(int)

# Merge basado en el ID
df_merged = pd.concat([df_pdb, df])

# Ordenar el DataFrame en función del ID
df_merged = df_merged.sort_values('ID')

# Reiniciar los índices del DataFrame
df_merged = df_merged.reset_index(drop=True)
```

```python
df_merged['code'] = df_merged['Aminoacids'].map(aminoacid_codification)
df_merged
```

```python
# Obtener la secuencia de aminoácidos
secuencia = ''.join(df_merged['code'][107:])

# Dividir la secuencia cada 75 aminoácidos
secuencia_dividida = [secuencia[i:i+75] for i in range(0, len(secuencia), 75)]

# Mostrar los segmentos de la secuencia dividida
for segmento in secuencia_dividida:
    print(segmento)
```
